# Projeto Python IA: Inteligência Artificial e Previsões

### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

Arquivos da aula: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
# Analisando 1 cliente
# se o cliente é Bom,OK ou Ruim (Nota de Crédito dele = Score de Crédito)

# Good = Boa
# Poor = Ruim
# Standard = OK

# Passo 0: Entender o desafio da empresa
# Passo 1: Importar a base de dados
import pandas as pd
base = pd.read_csv("clientes.csv")

display(base)
display(base.info())

# int -> números inteiros
# float -> números decimais
# object -> texto
```

```python
# Passo 2: Tratar e preparar a base de dados para a inteligência artificial

# profissao - cientista, marceneiro, pedreiro, pintor, artista, mecanico

from sklearn.preprocessing import LabelEncoder


# LabelEncoder -> transformar texto em número

# cientista = 1
# marceneiro = 2
# pedreiro = 3
# pintor = 4
# artista = 5
# mecanico = 6

codificador_profissao = LabelEncoder()
base["profissao"] = codificador_profissao.fit_transform(base["profissao"])
# mix_credito
codificador_mix = LabelEncoder()
base["mix_credito"] = codificador_mix.fit_transform(base["mix_credito"])
# comportamento_pagamento
codificador_comportamento = LabelEncoder()
base["comportamento_pagamento"] = codificador_comportamento.fit_transform(base["comportamento_pagamento"])

# quero prever a coluna "Score de Crédito" (variável alvo)
display(base)
```

```python

# y = é quem eu vou prever
y = base["score_credito"]

# x = quem eu vou usar para fazer a previsão
x = base.drop(columns=["score_credito", "id_cliente"])

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
# Passo 3: Criar o modelo de Previsão ou Modelo de IA -> Ruim,OK,Boa
# Arvore de decisão = RandomForest
# KMN = Nearest Neighbors - Vizinhos Próximos

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

modelo_arvoredecisao = RandomForestClassifier()
modelo_kmn = KNeighborsClassifier()

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_kmn.fit(x_treino, y_treino)
```

```python
# Passo 4: Avaliar e Escolher o melhor modelo de IA
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_kmn = modelo_kmn.predict(x_teste)

from sklearn.metrics import accuracy_score

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_kmn))
```

```python
# Passo 5: Fazer novas previsões

# melhor modelo = Arvore de Decisão
# quero prever o score de crédito de um novo cliente

novos_clientes = pd.read_csv("novos_clientes.csv")

display(novos_clientes)

novos_clientes["profissao"] = codificador_profissao.transform(novos_clientes["profissao"])
novos_clientes["mix_credito"] = codificador_mix.transform(novos_clientes["mix_credito"])
novos_clientes["comportamento_pagamento"] = codificador_comportamento.transform(novos_clientes["comportamento_pagamento"])     

previsao = modelo_arvoredecisao.predict(novos_clientes)
display(previsao)
```
